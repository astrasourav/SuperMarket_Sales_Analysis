```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/kaggle/input/supermarket-sales/supermarket_sales - Sheet1.csv')
```

```python
df.head()
# df.info()
# df.describe()
```

```python
# checking for duplicate valaues
duplicate = df.duplicated().sum()
duplicate
```

```python
# Preparing Data:Extract and count the Gender of consumers

consumer = df['Gender'].value_counts()
consumer
```

```python
# (Pie Chart)
plt.figure(figsize=(10,8))
consumer.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['blue', 'red'])

plt.title('Gender Type', fontsize=16)
plt.ylabel('')

plt.show()
```

```python
# (Bar Chart)
plt.figure(figsize=(10,8))
consumer.plot(kind='bar', color='red')

plt.title('Gender of consumer', fontsize=13)
plt.xlabel('Gender', fontsize=10)
plt.ylabel('Count', fontsize=10)

plt.show()
```

```python
# Now we will try to the most used mode of payment by the consumers

payment = df['Payment'].value_counts()
payment
```

```python
# Now plot pie chart 
plt.figure(figsize=(10, 8))
payment.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['blue', 'green', 'yellow'])

plt.title('Modes of payment', fontsize=(13))
plt.ylabel('')

plt.show()
```

```python
# Now we will find the type of customer
customerType = df['Customer type'].value_counts()
customerType
```

```python
# Now plot pie chart 
plt.figure(figsize=(10, 8))
customerType.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['blue', 'green'])

plt.title('Customer Type', fontsize=(13))
plt.ylabel('')

plt.show()
```

```python
# Take out Product Category
category = df['Product line'].value_counts()
category
```

```python
# (Pie Chart)
plt.figure(figsize=(10, 8))
category.plot(kind='bar', color='blue')

plt.title('Product Category', fontsize=13)
plt.xlabel('Products', fontsize=10)
plt.ylabel('Counts', fontsize=10)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

plt.show()
```

```python
# Now plot pie chart 
plt.figure(figsize=(10, 8))
category.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['blue', 'green', 'red', 'yellow', 'orange', 'brown'])

plt.title('Product Type', fontsize=(13))
plt.ylabel('')

plt.show()
```

```python
# Sales by Product Category
sales = df.groupby('Product line')['Total'].sum()
sales
```

```python
# Now plot pie chart 
plt.figure(figsize=(10, 8))
sales.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['blue', 'green', 'red', 'yellow', 'orange', 'brown'])

plt.title('Sales by Product Category', fontsize=(13))
plt.ylabel('')

plt.show()
```

```python
# Create a bar plot for product line sales
plt.figure(figsize=(10, 6))
sales.plot(kind='bar', color='skyblue')

# Add titles and labels
plt.title('Total Sales by Product Line', fontsize=16)
plt.xlabel('Product Line', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Total Sales by City
citySales = df.groupby('City')['Total'].sum()
citySales
```

```python
# Bar Chart
plt.figure(figsize=(10, 8))
citySales.plot(kind='bar', color='lightblue')

plt.title('City Sales', fontsize=13)
plt.xlabel('City Name', fontsize=10)
plt.ylabel('Total Amount', fontsize=10)

plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

```python
# Now plot pie chart 
plt.figure(figsize=(10, 8))
citySales.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['blue', 'green', 'red'])

plt.title('City Sales', fontsize=(13))
plt.ylabel('')

plt.show()
```

```python
# Average Transaction Size
avg_trans_by_gender = df.groupby('Gender')['Total'].mean()
avg_trans_by_gender
```

```python
# Bar Chart
plt.figure(figsize=(10, 8))
avg_trans_by_gender.plot(kind='bar', color='lightblue')

plt.title('Average Transaction Size (by Gender)', fontsize=13)
plt.xlabel('Gender', fontsize=10)
plt.ylabel('Total Amount', fontsize=10)

plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

```python
avg_trans_by_product = df.groupby('Product line')['Total'].mean()
avg_trans_by_product
```

```python
# Bar Chart
plt.figure(figsize=(10, 8))
avg_trans_by_product.plot(kind='bar', color='lightblue')

plt.title('Average Transaction Size (by Product)', fontsize=13)
plt.xlabel('Product', fontsize=10)
plt.ylabel('Total Amount', fontsize=10)

plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```
